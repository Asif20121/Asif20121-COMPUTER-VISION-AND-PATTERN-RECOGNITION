```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data2.csv')
df.head()

x1 = df['x1'].values.tolist()
x2 = df['x2'].values.tolist()
y = df['y'].values.tolist()

for i in range(len(y)):
    if y[i] == 0:
        plt.plot(x1[i], x2[i], 'r+')
    else:
        plt.plot(x1[i], x2[i], 'g+')
```

```python
import random
import math
from IPython import display

w = [random.random(), random.random()]
b = random.random()
lr = 0.0001
theta = 0.5
epoch = 0
count = 0

while count < len(y):
    total_error = 0
    for i in range(len(y)):
        ws = w[0]*x1[i] + w[1]*x2[i] + b
        pred = 0
        if ws > theta:
            pred = 1
            
        error = y[i] - pred
        if error != 0:
            count = 0
            total_error += math.sqrt(error**2)
            w[0] += error*x1[i]*lr
            w[1] += error*x2[i]*lr
            b += error*lr
        else:
            count += 1
            
    plt.figure()
    for i in range(len(y)):
        if y[i] == 0:
            plt.plot(x1[i], x2[i], 'r+')
        else:
            plt.plot(x1[i], x2[i], 'g+')
        
        plt.xlim([1,100])
        plt.ylim([1,100])
        
        xint = (theta-b)/ w[0]
        yint = (theta-b)/ w[1]
        slope = -yint/xint
        X = plt.xlim()
        Y = []
        for i in X:
            Y.append(slope*i + yint)
        plt.plot(X,Y, 'b-')
        
    plt.show()
```
